decision tree with iris data set

```python
import sklearn.datasets as datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
iris
```

```python
type(iris.data)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
y = iris.target
y
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(df, y)
```

```python
dtree.predict([
    [1.1, 1.8, 1.2, 0.2],
    [8.1, 2.1, 2.5, 2.9]
])
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Ensure this list matches the model's understanding of classes
class_names = ['class1', 'class2', 'class3']  # Update based on your model's classes

plt.figure(figsize=(20, 10))
tree.plot_tree(dtree,
               filled=True,
               rounded=True,
               feature_names=df.columns,
               class_names=class_names,  # Make sure this is correct
               fontsize=12)
plt.show()
```

Decision trees utilize two key metrics for defining the splits at each node: entropy and the Gini index. Initially, entropy is calculated for each column of data to measure the impurity (or disorder) within the data. The column with the highest entropy is selected as the root of the tree because it has the greatest potential to improve classification. Subsequently, for each data sample in this column, the Gini index is calculated. The sample that results in the minimum Gini index (i.e., the least impurity after the split) is chosen as the condition for the new node. This process assists in data segmentation and constructs part of the tree structure to optimize decision-making in predictions.
